```python
# Initialize the WebDriver
driver = webdriver.Chrome()

# Load the webpage
driver.get("https://www.redbus.in/online-booking/ksrtc-kerala/?utm_source=rtchometile")
time.sleep(3)
driver.maximize_window()

# Function to retrieve bus links and routes
def Kerala_link_route(path):
    LINKS_KERALA = []
    ROUTE_KERALA = []
    wait = WebDriverWait(driver, 20)
    
    for i in range(1, 3):
        paths = driver.find_elements(By.XPATH, path)
        
        # Retrieve the route links
        for link in paths:
            LINKS_KERALA.append(link.get_attribute("href"))
        
        # Retrieve the names of the routes
        for route in paths:
            ROUTE_KERALA.append(route.text)
        
        try:
            # Find the pagination button using a more general approach
            pagination_buttons = driver.find_elements(By.XPATH, '//div[@class="DC_117_pageTabs "]')
            if i < len(pagination_buttons):
                next_button = pagination_buttons[i]
                driver.execute_script("arguments[0].scrollIntoView();", next_button)
                time.sleep(2)
                next_button.click()
            else:
                print(f"No more pages to paginate at step {i}")
                break
        
        except (NoSuchElementException, ElementClickInterceptedException) as e:
            print(f"Exception encountered: {e}. No more pages to paginate at step {i}")
            break
    
    return LINKS_KERALA, ROUTE_KERALA

LINKS_KERALA, ROUTE_KERALA = Kerala_link_route("//a[@class='route']")

# Create a DataFrame with the retrieved data
df_k = pd.DataFrame({"Route_name": ROUTE_KERALA, "Route_link": LINKS_KERALA})
print(df_k)

# Close the WebDriver
driver.quit()
```
